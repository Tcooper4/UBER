```python
# Test 2: read all TLC files with schema handling
result = duckdb.query("""
    SELECT COUNT(*) as rows, MIN(pickup_datetime) as earliest, MAX(pickup_datetime) as latest
    FROM read_parquet('data/raw/tlc/*.parquet', union_by_name=True)
""").df()
print(result)
```
